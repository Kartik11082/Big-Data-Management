# Friend Recommendation using Mutual Friends

Read the Data

```
DATA_PATH = "dbfs:/FileStore/tables/assignment2_data.txt"
```

```
raw_data = sc.textFile(DATA_PATH)
```

```
raw_data.take(3)
```

Parse the data

```
def parse_line(line):
    parts = line.split("\t")
    if len(parts) != 2:
        return None
    user = parts[0]
    friends = parts[1].split(",")
    return (user, friends)
```

```
user_friends = raw_data.map(parse_line).filter(lambda x: x is not None)
```

```
user_friends.take(2)
```

```
user_friends_dict = user_friends.collectAsMap()
```

```
user_friends_dict.items()
```

```
broadcast_friends = sc.broadcast(user_friends_dict)
```

```
type(broadcast_friends)
```

```
broadcast_friends.value.get("0")
```

Generate Pairs of friends

```
# Generate candidate pairs using mutual friends.
# For each user, emit direct friendship markers and candidate pairs.
def generate_pairs(user_friends_tuple):
    user, friends = user_friends_tuple
    pairs = []
    # Emit direct friendship markers: these will have value 0.
    for friend in friends:
        pairs.append(((user, friend), 0))
    # For each friend of the user, look up their friends and emit candidate pairs with value 1.
    for friend in friends:
        friend_friends = broadcast_friends.value.get(friend, [])
        for candidate in friend_friends:
            if candidate != user:  # Avoid self-loops
                pairs.append(((user, candidate), 1))
    return pairs
```

```
# Use flatMap to generate all candidate pairs.
candidate_pairs = user_friends.flatMap(generate_pairs)
```

```
candidate_pairs.take(5)
```

Sum the mutual friend counts for each candidate pair.

```
mutual_counts = candidate_pairs.reduceByKey(lambda a, b: a + b)
```

```
mutual_counts.take(10)
```

Filter out pairs that are already friends.
- For a direct friendship, we expect the value to be 0 (because we added a marker).
- Only candidates with mutual friend count > 0 are recommendations.


```
recommendations = mutual_counts.filter(lambda x: x[1] > 0).map(
    lambda x: (x[0][0], (x[0][1], x[1]))
)
```

```
recommendations.take(5)
```

Group recommendations by the user and sort the candidate list to pick top 10.

```
recommendations_grouped = recommendations.groupByKey().mapValues(
    lambda recs: sorted(list(recs), key=lambda y: -y[1])[:10]
)
```

Convert to DataFrame and display a random sample of 10 users’ recommendations.

```
rec_df = recommendations_grouped.toDF(["User", "Recommendations"])
sample_recs = rec_df.sample(withReplacement=False, fraction=0.05).limit(10)
```

```
sample_recs.filter(sample_recs["user"] == "0").display()
```
